```python
binary_columns = ["Smoking","FamilyHistoryAlzheimers","CardiovascularDisease","Diabetes",
                  "Depression","HeadInjury","Hypertension","MemoryComplaints","BehavioralProblems",
                 "Confusion","Disorientation","PersonalityChanges","DifficultyCompletingTasks",
                 "Forgetfulness","Diagnosis"]
```

```python
mapping_binary_columns = {0: "No", 1: "Yes"}
```

```python
df[binary_columns] = df[binary_columns].replace(mapping_binary_columns)
```

```python
df.head()
```

```python
mapping_gender = {0: "Male", 1: "Female"}
mapping_ethnicity = {0: "Caucasian", 1: "African American", 2: "Asian", 3: "Other"}
mapping_educationlevel = {0: "None", 1: "High School", 2: "Bachelor's", 3: "Higher"}
```

```python
df["Gender"] = df["Gender"].replace(mapping_gender)
df["Ethnicity"] = df["Ethnicity"].replace(mapping_ethnicity)
df["EducationLevel"] = df["EducationLevel"].replace(mapping_educationlevel)
```

```python
df.head()
```

```python
categorical_features = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking',
 'FamilyHistoryAlzheimers', 'CardiovascularDisease',
'Diabetes', 'Depression', 'HeadInjury', 'Hypertension',
'MemoryComplaints', 'BehavioralProblems', 'Confusion', 'Disorientation',
'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness', 'Diagnosis']
```

```python
numeric_features = [col for col in df.columns if col not in categorical_features]
```

```python
numeric_features
```

```python
df[numeric_features].head()
```

```python
float_features = ['BMI',
 'AlcoholConsumption',
 'PhysicalActivity', 
 'DietQuality', 
 'SleepQuality',
 'CholesterolTotal',
 'CholesterolLDL',
 'CholesterolHDL',
 'CholesterolTriglycerides',
 'MMSE',
 'ADL',
 'FunctionalAssessment']
```

```python
df.isnull().sum()
```

```python
df[float_features] = df[float_features].round(2)
```
